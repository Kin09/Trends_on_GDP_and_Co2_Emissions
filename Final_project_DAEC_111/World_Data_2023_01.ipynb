```python
import pycountry_convert

# Function to get continent name from country name
def get_continent(country):
    try:
        country_code = pycountry_convert.country_name_to_country_alpha2(country)
        continent_code = pycountry_convert.country_alpha2_to_continent_code(country_code)
        continent_name = pycountry_convert.convert_continent_code_to_continent_name(continent_code)
        return continent_name
    except:
        return ''

# Add 'Continent' column to the dataset
df['Continent'] = df['Country'].apply(get_continent)

# Print the updated dataset
print(df['Continent'])
```

```python
# Filter the dataset to include relevant columns
df_filtered = df[['Country', 'Continent']]

# Group countries by continent
grouped_countries = df_filtered.groupby('Continent')

# Iterate over groups
for continent, countries in grouped_countries:
    # Display continent and countries
    print(continent)
    print(countries['Country'].values)
    print()
```

```python
# Replace missing values with 0 for numeric columns
numeric_columns = ['Birth Rate', 'Calling Code', 'Fertility Rate', 'Infant mortality', 'Life expectancy', 'Maternal mortality ratio', 'Latitude', 'Longitude']
df[numeric_columns] = df[numeric_columns].fillna(0)

# Replace missing values with an empty string for object columns
object_columns = ['Country', 'Density\n(P/Km2)', 'Abbreviation', 'Agricultural Land( %)', 'Land Area(Km2)', 'Armed Forces size', 'Capital/Major City', 'Co2-Emissions', 'CPI', 'CPI Change (%)', 'Currency-Code', 'Forested Area (%)', 'Gasoline Price', 'GDP', 'Gross primary education enrollment (%)', 'Gross tertiary education enrollment (%)', 'Largest city', 'Minimum wage', 'Official language', 'Unemployment rate', 'Urban_population']
df[object_columns] = df[object_columns].fillna('')

# Check if missing values have been handled
print(df.isnull().sum())
```
